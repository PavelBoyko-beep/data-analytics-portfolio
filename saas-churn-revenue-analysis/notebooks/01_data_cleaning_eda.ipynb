```python
missing_values = pd.DataFrame()

for table_name, df in tables.items():

    missing_count = df.isnull().sum()

    missing_percent = (
        df.isnull().sum() / len(df) * 100
    ).round(2)

    temp_df = pd.DataFrame({
        "column_name": missing_count.index,
        "missing_count": missing_count.values,
        "missing_percent": missing_percent.values
    })

    temp_df["table_name"] = table_name

    missing_values = pd.concat(
        [missing_values, temp_df],
        ignore_index=True
    )

missing_values = (
    missing_values
    .sort_values(
        by="missing_count",
        ascending=False
    )
)

missing_values.head(20)
```

```python
duplicate_rows = pd.DataFrame()

for table_name, df in tables.items():

    duplicate_count = df.duplicated().sum()

    temp_df = pd.DataFrame({
        "table_name": [table_name],
        "duplicate_rows": [duplicate_count],
        "total_rows": [len(df)]
    })

    temp_df["duplicate_percent"] = (
        temp_df["duplicate_rows"] / temp_df["total_rows"] * 100
    ).round(2)

    duplicate_rows = pd.concat(
        [duplicate_rows, temp_df],
        ignore_index=True
    )

duplicate_rows
```

## Primary Key Check Notes

The primary key check found that `usage_id` in the `feature_usage` table is not fully unique.

- Total rows in `feature_usage`: 25,000
- Unique `usage_id` values: 24,979
- Duplicate `usage_id` values: 21
- Missing `usage_id` values: 0

The duplicated `usage_id` values belong to different usage events, so these rows should not be removed.

Decision:
For SQL import, `usage_id` will be kept as the original raw identifier, but it will not be used as the primary key. A technical primary key should be created for the `feature_usage` table during PostgreSQL setup.

```python
foreign_key_checks = []

# 1. subscriptions.account_id should exist in accounts.account_id
missing_subscription_accounts = subscriptions[
    ~subscriptions["account_id"].isin(accounts["account_id"])
]

foreign_key_checks.append({
    "relationship": "subscriptions.account_id -> accounts.account_id",
    "missing_foreign_keys": len(missing_subscription_accounts)
})

# 2. support_tickets.account_id should exist in accounts.account_id
missing_ticket_accounts = support_tickets[
    ~support_tickets["account_id"].isin(accounts["account_id"])
]

foreign_key_checks.append({
    "relationship": "support_tickets.account_id -> accounts.account_id",
    "missing_foreign_keys": len(missing_ticket_accounts)
})

# 3. churn_events.account_id should exist in accounts.account_id
missing_churn_accounts = churn_events[
    ~churn_events["account_id"].isin(accounts["account_id"])
]

foreign_key_checks.append({
    "relationship": "churn_events.account_id -> accounts.account_id",
    "missing_foreign_keys": len(missing_churn_accounts)
})

# 4. feature_usage.subscription_id should exist in subscriptions.subscription_id
missing_usage_subscriptions = feature_usage[
    ~feature_usage["subscription_id"].isin(subscriptions["subscription_id"])
]

foreign_key_checks.append({
    "relationship": "feature_usage.subscription_id -> subscriptions.subscription_id",
    "missing_foreign_keys": len(missing_usage_subscriptions)
})

foreign_key_checks = pd.DataFrame(foreign_key_checks)

foreign_key_checks
```

## Feature Usage Date Limitation

The date logic checks found a major limitation in the `feature_usage` table.

Initial check:
- 19,142 usage events happened before the related subscription start date.
- This represents 76.57% of all feature usage records.

Additional validation:
- For 4,122 subscriptions, the first usage event happened before the related account signup date.
- This represents 82.44% of subscription records.

Interpretation:
This means `feature_usage.usage_date` should not be used for lifecycle sequencing, such as:
- time from signup to first usage;
- usage before/after subscription start;
- time-to-value analysis;
- cohort analysis based on first usage date.

Decision:
The `feature_usage` table will still be used for product activity analysis, such as feature usage volume, error counts, and beta feature usage. However, usage dates will not be used for business conclusions that depend on the order of signup, subscription start, and usage events.

```python
numeric_checks = []

# accounts
numeric_checks.append({
    "table_name": "accounts",
    "column_name": "seats",
    "negative_values": (accounts["seats"] < 0).sum(),
    "zero_values": (accounts["seats"] == 0).sum(),
    "min_value": accounts["seats"].min(),
    "max_value": accounts["seats"].max()
})

# subscriptions
for column in ["seats", "mrr_amount", "arr_amount"]:
    numeric_checks.append({
        "table_name": "subscriptions",
        "column_name": column,
        "negative_values": (subscriptions[column] < 0).sum(),
        "zero_values": (subscriptions[column] == 0).sum(),
        "min_value": subscriptions[column].min(),
        "max_value": subscriptions[column].max()
    })

# feature_usage
for column in ["usage_count", "usage_duration_secs", "error_count"]:
    numeric_checks.append({
        "table_name": "feature_usage",
        "column_name": column,
        "negative_values": (feature_usage[column] < 0).sum(),
        "zero_values": (feature_usage[column] == 0).sum(),
        "min_value": feature_usage[column].min(),
        "max_value": feature_usage[column].max()
    })

# support_tickets
for column in ["resolution_time_hours", "first_response_time_minutes"]:
    numeric_checks.append({
        "table_name": "support_tickets",
        "column_name": column,
        "negative_values": (support_tickets[column] < 0).sum(),
        "zero_values": (support_tickets[column] == 0).sum(),
        "min_value": support_tickets[column].min(),
        "max_value": support_tickets[column].max()
    })

# churn_events
numeric_checks.append({
    "table_name": "churn_events",
    "column_name": "refund_amount_usd",
    "negative_values": (churn_events["refund_amount_usd"] < 0).sum(),
    "zero_values": (churn_events["refund_amount_usd"] == 0).sum(),
    "min_value": churn_events["refund_amount_usd"].min(),
    "max_value": churn_events["refund_amount_usd"].max()
})

numeric_checks = pd.DataFrame(numeric_checks)

numeric_checks
```

### Key Findings

1. `subscriptions.end_date` has many missing values, but this is expected because active subscriptions do not have an end date.

2. `support_tickets.satisfaction_score` has missing values, but this is expected because not every customer leaves a satisfaction rating.

3. `churn_events.feedback_text` has missing values, but this is expected because not every churned customer provides written feedback.

4. The `feature_usage` table has 21 duplicated `usage_id` values. These rows are not full duplicates and should not be removed. For SQL import, a technical primary key should be created for the `feature_usage` table.

5. Foreign key checks passed. All subscriptions, support tickets, churn events, and feature usage records are linked to valid parent records.

6. Date columns were converted to proper datetime types.

7. The `feature_usage.usage_date` field has a lifecycle inconsistency: many usage records occur before the related account signup date or subscription start date. Because of this, `usage_date` will not be used for lifecycle sequencing or time-to-first-usage analysis.

8. Zero MRR/ARR values are explained by trial subscriptions. These rows should not be removed.

9. Numeric fields do not contain negative values.

10. Satisfaction scores are within the valid range from 1 to 5 when provided.

### Data Quality Decisions

- Keep missing `end_date` values because they represent active subscriptions.
- Keep missing `satisfaction_score` values because they represent customers who did not leave a rating.
- Keep zero MRR/ARR rows because they represent trial subscriptions.
- Keep all `feature_usage` rows, but do not use `usage_id` as the primary key.
- Do not use `feature_usage.usage_date` for lifecycle-based conclusions.
- Use `feature_usage` for product activity analysis: feature usage volume, error counts, and beta feature usage.

```python
# 8.2 Current MRR by Plan Tier

# Convert subscription dates to datetime format
subscriptions["start_date"] = pd.to_datetime(subscriptions["start_date"])
subscriptions["end_date"] = pd.to_datetime(subscriptions["end_date"])

# Use the latest month available in the subscription data
report_date = (
    subscriptions["start_date"].max()
    .to_period("M")
    .to_timestamp("M")
)

print("Report date:", report_date.date())

# Keep only active paid subscriptions at the report date
active_paid_subscriptions = subscriptions[
    (subscriptions["start_date"] <= report_date) &
    (
        subscriptions["end_date"].isna() |
        (subscriptions["end_date"] >= report_date)
    ) &
    (subscriptions["is_trial"] == False) &
    (subscriptions["mrr_amount"] > 0)
]

# Aggregate current MRR by plan tier
mrr_by_plan = (
    active_paid_subscriptions
    .groupby("plan_tier", as_index=False)
    .agg(
        active_paid_subscriptions=("subscription_id", "nunique"),
        current_mrr=("mrr_amount", "sum")
    )
    .sort_values("current_mrr", ascending=False)
)

# Add MRR in millions for easier chart reading
mrr_by_plan["current_mrr_millions"] = mrr_by_plan["current_mrr"] / 1_000_000

mrr_by_plan
```

```python
# 8.3 Chart: Current MRR by Plan Tier

plt.figure(figsize=(8, 5))

plt.bar(
    mrr_by_plan["plan_tier"],
    mrr_by_plan["current_mrr_millions"]
)

plt.title("Current MRR by Plan Tier")
plt.xlabel("Plan Tier")
plt.ylabel("Current MRR, millions")

plt.show()
```

```python
# 8.6 Support Escalation Rate by Churn Status

# Make sure escalation_flag is treated as True/False
support_tickets["escalation_flag"] = (
    support_tickets["escalation_flag"]
    .astype(str)
    .str.lower()
    .eq("true")
)

# Aggregate support metrics at account level
account_support = (
    support_tickets
    .groupby("account_id", as_index=False)
    .agg(
        support_tickets=("ticket_id", "count"),
        escalated_tickets=("escalation_flag", "sum"),
        avg_resolution_time_hours=("resolution_time_hours", "mean"),
        avg_first_response_minutes=("first_response_time_minutes", "mean"),
        avg_satisfaction_score=("satisfaction_score", "mean")
    )
)

# Count churn events at account level
account_churn = (
    churn_events
    .groupby("account_id", as_index=False)
    .agg(
        churn_events=("churn_event_id", "count")
    )
)

# Combine accounts with support and churn information
account_level = (
    accounts[["account_id"]]
    .merge(account_support, on="account_id", how="left")
    .merge(account_churn, on="account_id", how="left")
)

# Replace missing counts with 0
account_level["support_tickets"] = account_level["support_tickets"].fillna(0)
account_level["escalated_tickets"] = account_level["escalated_tickets"].fillna(0)
account_level["churn_events"] = account_level["churn_events"].fillna(0)

# Create churn status label
account_level["churn_status"] = account_level["churn_events"].apply(
    lambda x: "churned" if x > 0 else "not_churned"
)

# Summarize support metrics by churn status
support_churn_summary = (
    account_level
    .groupby("churn_status", as_index=False)
    .agg(
        accounts=("account_id", "nunique"),
        accounts_with_support_tickets=("support_tickets", lambda x: (x > 0).sum()),
        churn_events=("churn_events", "sum"),
        avg_support_tickets_per_account=("support_tickets", "mean"),
        total_support_tickets=("support_tickets", "sum"),
        total_escalated_tickets=("escalated_tickets", "sum"),
        avg_resolution_time_hours=("avg_resolution_time_hours", "mean"),
        avg_first_response_minutes=("avg_first_response_minutes", "mean"),
        avg_satisfaction_score=("avg_satisfaction_score", "mean")
    )
)

# Calculate escalation rate
support_churn_summary["escalation_rate_percent"] = (
    support_churn_summary["total_escalated_tickets"]
    / support_churn_summary["total_support_tickets"]
    * 100
).round(2)

# Round numeric columns for easier reading
support_churn_summary = support_churn_summary.round(2)

support_churn_summary
```

```python
# 8.7 Chart: Support Escalation Rate by Churn Status

plt.figure(figsize=(7, 5))

plt.bar(
    support_churn_summary["churn_status"],
    support_churn_summary["escalation_rate_percent"]
)

plt.title("Support Escalation Rate by Churn Status")
plt.xlabel("Churn Status")
plt.ylabel("Escalation Rate, %")

plt.show()
```

### Chart 3 — Support Escalation Rate by Churn Status

This chart compares support escalation rate between churned and non-churned accounts.

Churned accounts have a slightly higher escalation rate: **5.09%**, compared with **3.97%** for non-churned accounts.

This supports the SQL finding that churned accounts show a somewhat higher support escalation rate, while other support metrics are very similar.

**Important limitation:**
